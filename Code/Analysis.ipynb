```python
df = df[['coaster_name', 'Manufacturer', 'Opened', 'year_introduced', 'latitude', 'longitude', 'Type_Main', 'opening_date_clean', 'speed_mph', 'height_ft','Inversions_clean', 'Gforce_clean', 'Location', 'Status']].copy()
```

```python
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
```

```python
#Rename our columns

df = df.rename(columns={"coaster_name" : "Coaster_Name",
                   "year_introduced" : "Year_Introduced",
                   "opening_date_clean" : "Opening_Date",
                   "speed_mph" : "Speed_mph",
                   "height_ft" : "Height_ft",
                   "Inversions_clean" : "Inversions",
                   "Gforce_clean" : "Gforce"
                })
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
#identify Duplicates

df = df.drop_duplicates(subset=["Coaster_Name", "Location", "Opening_Date"]).reset_index(drop=True)
```

```python
df["Year_Introduced"].value_counts()
```

```python
ax = df["Year_Introduced"].value_counts().head(10).plot(kind="barh", title="Top 10 Years Coasters Introduced")
ax.set_xlabel("Count of Coaster")
ax.set_ylabel("Year")
```

```python
ax = df["Speed_mph"].plot(kind="hist", bins=30, title="Coaster Speed (mph)")
ax.set_xlabel("Speed (mph)")
```

```python
ax = df.plot(kind="scatter", x="Speed_mph", y="Height_ft", title="Coster speed VS Height")
ax.set_xlabel("Speed")
ax.set_ylabel("Height")
```

```python
sns.scatterplot(x="Speed_mph", y="Height_ft", data=df, hue="Year_Introduced")
```

```python
sns.pairplot(data=df, vars=["Year_Introduced", "Speed_mph", "Height_ft", "Inversions", "Gforce"])
```

```python
df_corr = df[["Year_Introduced", "Speed_mph", "Height_ft", "Inversions", "Gforce"]].dropna().corr()
```

```python
sns.heatmap(df_corr, annot = True, linewidths=1, linecolor="white")
```

```python
ax = df.query('Location != "Other"').groupby("Location")["Speed_mph"].agg(["mean", "count"]).query('count >= 10').sort_values("mean")["mean"].plot(kind="barh", figsize=(12,5), title="Average Coast Speed by Location")
ax.set_xlabel("Average Coaster Speed")
```
